```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

```python
observations=1000

xs=np.random.uniform(-10,10,size=(observations,1))
zs=np.random.uniform(-10,10,size=(observations,1))

generated_inputs=np.column_stack((xs,zs))
#inputs.shape

noise =np.random.uniform(-1,1,size=(observations,1))
generated_targets = 13*xs + 7*zs - 12 + noise

np.savez('Tf_intro', input =generated_inputs,targets=generated_targets)
```

```python
training_data= np.load('Tf_intro.npz')
```

```python
input_size=2
output_size=1

#model= tf.keras.Sequential([tf.keras.layers.Dense(output_size)])
model=tf.keras.Sequential([tf.keras.layers.Dense(output_size,kernel_initializer=tf.random_uniform_initializer(minval=-0.1,maxval=0.1)
                                                ,bias_initializer=tf.random_uniform_initializer(minval=-0.1,maxval=0.1))])

#model.compile(optimizer='sgd',loss='mean_squared_error')
custom_optimizer=tf.keras.optimizers.SGD(learning_rate=0.02)
#model.compile(optimizer=custome_optimizer,loss='mean_squared_error') Mean_squared_error is L2 norm loss normalized based on Observations

model.compile(optimizer=custom_optimizer,loss='huber_loss')
model.fit(training_data['input'],training_data['targets'],epochs=100, verbose=2)
```

Why do we need non linearities? 

Ans . To stack layers in a deep net. If only a linear relationship exists, even if we have 100 layers it can be simplified
to a single linear relationship . So inorder to stack different layers and  find complex relationships through arbitrary 
functions we need non-linearities

Non linearity = Activation or transfer functions

    1.sigmoid or logistic function(derivative of the sigmoid formula is calculated automatically by tf)
    2.TanH or hyperbolic tangent
    3.ReLu- rectified linear unit
    4.softmax- graph is different everytime- each element depends on the entire set of elements(sigma of all elements taken in deno)
sum of values of outputs is always equal to one - So the softmax has a connection with probability

    The softmax transformation transforms a bunch of arbitrary large or small nos into a valid probablity distribution

    Softmax is used as the activation func for the output layer in classification problems

# Intialization

Process by which we set the initial value of weights
iF  weights are not intialized properly - optimization would get affected

The weights should be random as well otherwise when in a nueral network the model wont have any reason to think that the 
different nodes of the hidden layers are different( or to think even the outputs are diff)

Initialization types - Normal or uniform 

Make sure that the initializatio has a pretty decent range from where it is derived. Why?

When applying activation functions like sigmoid, if the values are small (around 0) the sigmoid function is actually
linear around this area and therefore non linearity wont happen.

    Initialization used in tf - Xavier Initialization
    
    Uniform Xavier initialization - weight to be drawn from a uniform dist [-x,x] where x = sqrt(6/no of inputs+outputs) 
    Normal Xavier Initialization- weight to be drawn from a normal dist with mean 0 and std dev, 
    sigma = sqrt(2/no of inputs+outputs) 
    
    The range should be based on the number of outputs in the following layer which is addressed by xavier Init

# OPTIMZATION

# SGD, GD and MOMENTUM

SGD and GD - SGD(Stocastic Gradient Descent) - the data set is split into batches. In GD the weight updates after 
each epoch (entire set)
whereas in SGD this happens after each batch. Suppose the data set is split into 10 batches, the weight updation in SGD
happens ten times when it happens only once in Gradient Descent

BUT SGD and GD is prone to error. If a local minima if found, the SGD and GD will accept it
which results in a sub optimal solution and miss the global minima.

For this we introduce a concept called MOMENTUM.when momentum is introduced the alogorithm falls into global min.
Similar to the case of a ball rolling downhill - If the descent is steep and the dip is shallow(local minima) the ball tends to roll
over due to its momentum and will stop only when the dip is deep enough.

MOMENTUM is inlcluded as follows : usual update step : w'= w- derivative of loss func with respect to weight
        MOMENTUM added :  w'=w(t)- partial derivate wrt weight at t-th time - alpha*partial derivative at t-1 th time
        
        alpha= 0.9(usually) - done so that different weightages are attached to the deivatives at t and t-1
            



# Learning Rate Schedules

High learning rates =  we wont reach the optimal soln, it either oscillates or deviates from the solution
low learning rates = we will reach the opt soln very slowly

so a good learning rate is very imperative and this can be acheived using a learning rate schedule

A learning rate schedule begins with a high learning rate value and decreases it as we proceed

A exponential learning rate schedule is highly effective. This method updates the learning rate value as follows
Learning rate = L.R(o) * e^(-n/c)

where n = current epoch
and C = decay coeff(~20) should be of the order of the epoch


Different types of learning rate schedules

1.Adagrad - Adaptive gradient algorithm
2.RMSprop - Root mean squared propogation
3.ADAM - Adaptive moment estimation

Adagrad - at the weight  update step, the learning rate is divided by a monotonously increasing value (G) {G+epsilon}
so the learning rate keeps decreasing 
RMS prop - same way as adagrad but the G value is not staeding increasing as a weight is multipled to the step which updates G

G updation in adagrad = G(t) = G(t-1) + square of partial derivated of loss func wrt weight ( so G is an increasing value)
G in RMSprop = G(t) = beta*G(t-1) + (1-beta)* square of partial derivative of loss func wrt weight ( G is not increasing due to the weight)


ADAM = Best method ( combines RMS prop and MOMENTUM ideas)

    ADAM or adaptive moment estimation is the best Optimizing method

    w(t) = w(t-1) - L.Rate/(G(t)+epsilon) * Mi(t)

    where Mi(t) = alpha*Mi(t-1) + (1-alpha)* partial derivative of loss function wrt Weight

# Preprocessing categorical data

Types - One Hot Encoding and Binary Encoding 
    
    Binary encoding - take binary values and split them into different variables (has problems - correlation issues)
    One hot encoding - widely adopted method (one major problem - no of variables/categories)
    
  if we have three variables - we make three columns( as many columns as the no of variables)
  Now we assign 1 or 0 based on whether it is the product under consideration or not 
    
  eg: Bread,yogurt,drink(3 by 3 marix)
  Bread would be (1,0,0)
  yogurt would be (0,1,0) and so on
    
    Cases where the number of categories is high ( like 1000 items) by one hot encoding method we would need to introduce 
    1000 variable sin our input whereas using binary, we can tone thsi down to something around 16 . So when the no of categories is high go with Binary method

```python
mnist_train,mnist_test=mnist_dataset['train'],mnist_dataset['test']
```

```python
mnist_test
```

```python
num_validation_samples = 0.1*mnist_info.splits['train'].num_examples
num_validation_samples=tf.cast(num_validation_samples,tf.int64)
```

```python
num_test_samples = mnist_info.splits['test'].num_examples
num_test_samples = tf.cast(num_test_samples,tf.int64)
```

# Preprocessing

```python
def scale(image,label):
    image=tf.cast(image,tf.float32)
    image/=255.                    # to keep as float
    return image,label            # return image and label since the dataset has these properties

scaled_train_and_validation_set =mnist_train.map(scale)   #.map() allows us to use a custom transformation

test_data = mnist_test.map(scale)

```

```python
#Shuffle the dataset

BUFFER_SIZE=10000
shuffled_train_and_validation=scaled_train_and_validation_set.shuffle(BUFFER_SIZE)
```

```python
#extracting validation and train data

validation_data = shuffled_train_and_validation.take(num_validation_samples)
train_data=shuffled_train_and_validation.skip(num_validation_samples)

#getting ready for optimizing( defining batch size for mini batch  gradient)

BATCH_SIZE=100
train_data = train_data.batch(BATCH_SIZE)
validation_data = validation_data.batch(num_validation_samples)

test_data = test_data.batch(num_test_samples)

validation_inputs,validation_targets =next(iter(validation_data))
```

# Outline the model

```python
input_layer=784
hidden_layer=150  #50 is the depth and two is the width chosen for the nueral net, we can play around with these nos to get better results
output_layer=10

model=tf.keras.Sequential([
        
        #first flatten the tensor into vector and then stack it
        
        tf.keras.layers.Flatten(input_shape=(28,28,1)),
        tf.keras.layers.Dense(hidden_layer,activation='relu'),
        tf.keras.layers.Dense(hidden_layer,activation='relu'),
        tf.keras.layers.Dense(hidden_layer,activation='tanh'),
        tf.keras.layers.Dense(hidden_layer,activation='tanh'),
        tf.keras.layers.Dense(output_layer,activation='softmax') # since the target is categorical variable and we would expect outoput as probabilities
                              
    
])
```

# Select the loss and optimizer

```python
#we will use the .compile for this

#custom_optimizer=tf.keras.optimizers.Adam(learning_rate=0.02)
#model.compile(optimizer=custom_optimizer,loss='sparse_categorical_crossentropy',metrics=['accuracy'])

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])

#cross entropy loss functions can be several types
#Binary_crossentropy when binary encoding is involved for our data
#categorical_crossentropy and sparse_categorical_crossentropy with diff being later is used when data is one-hot encoded which is the case here
```

# Training

```python
NUMBER_OF_EPOCHS =5

#fit the model

model.fit(train_data,epochs=NUMBER_OF_EPOCHS,validation_data=(validation_inputs,validation_targets),verbose=2)
```

# TEST THE MODEL

```python
test_loss,test_accuracy=model.evaluate(test_data)
```

```python
print('test loss : {0:.2f}. Test accuracy: {1:.2f}%'.format(test_loss,test_accuracy*100.))
```
